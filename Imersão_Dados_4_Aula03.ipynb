```python
dados_vendas.sample(3)
```

```python
dados_vendas['Valor_m2'] = dados_vendas['Valor_anuncio_float']/dados_vendas['Metragem']
# √© feito o valor do m2 de cada im√≥vel
dados_vendas.sample(3)
```

```python
dados_vendas.groupby('Bairro').mean().sample(3)
# essa m√©dia considera o valor por m2 de cada im√≥vel individualmente
```

```python
# para dar maior exatid√£o ao valor do m2 por bairro, o ideal √© somar todos os valores do bairro, depois todas as metragens do bairro
# depois dividir o valor pelo m2
dados_bairro = dados_vendas.groupby('Bairro').sum()
dados_bairro.sample(3)
```

```python
dados_bairro['Valor_m2_bairro'] = dados_bairro['Valor_anuncio_float']/dados_bairro['Metragem']
dados_bairro.sample(3)
```

```python
top_bairros = dados_vendas['Bairro'].value_counts()[:10].index
top_bairros
# retorna os 10 nomes de bairros que mais se repetem
```

```python
# dados_vendas.reset_index() # faz com que a coluna index seja renomeada para o "padr√£o" (nomes de bairro s√£o substituidos "0,1,...n")
dados_bairro.reset_index(inplace = True) # transforma a coluna index em uma coluna normal e insere um index "padr√£o"
```

```python
dados_bairro
```

```python
dados_bairro.query('Bairro in @top_bairros')
```

```python
plt.figure(figsize=(10, 5))
ax = sns.barplot(x='Bairro', y='Valor_m2_bairro', data=dados_bairro.query('Bairro in @top_bairros'))
ax.tick_params(axis='x', rotation=65, pad=-3) # "tick" s√£o os labels abaixo do gr√°fico / "pad" √© a dist√¢ncia entre "tick" e gr√°fico
```

```python
plt.figure(figsize=(10, 8))
ax = sns.boxplot(data = dados_vendas.query('Bairro in @top_bairros'), x='Bairro', y='Valor_anuncio_float')
ax.tick_params(axis='x', rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
ax = sns.boxplot(data = dados_vendas.query('Bairro in @top_bairros & Metragem < 2500'), x='Bairro', y='Metragem')
# tinha uma metragem de 750000 que distorcia muito o gr√°fico o gr√°fico, e outro com metragem de 2500 que tb era muito outlier
ax.tick_params(axis='x', rotation=45)
plt.show()
```
