```python
enderecos_sp = enderecos.query('sigla_uf == "SP"')
enderecos_sp.head(3)
```

```python
enderecos_sp['rua'] = enderecos_sp['tipo_logr'] + ' ' + enderecos_sp['logr_nome']
enderecos_sp['rua'] = enderecos_sp['rua'].str.lower().str.strip()
enderecos_sp.head(3)
```

```python
dados_vendas['Rua'].str.extract(r'(^[\w ]+)')[:10] # regex (expressão regular)
```

```python
dados_vendas['apenas_rua'] = dados_vendas['Rua'].str.extract(r'(^[\w ]+)') # esse regex descarta o ", num" de cada valor da coluna "Rua"
dados_vendas['apenas_rua'] = dados_vendas['apenas_rua'].str.lower().str.strip() # deixa tudo em minúsculo e tira espaços extras
#isso deixa os endereços com o mesmo padrão de formatação da base "enderecos_sp"
dados_vendas.head(3)
```

```python
dados_geo = pd.merge(left = dados_vendas, right = enderecos_sp[['rua', 'cep', 'latitude', 'longitude']], how = 'left', left_on = 'apenas_rua', right_on = 'rua')
# "dados_geo" vai ser um novo DB
# o "merge()" vai unir as duas tabelas. É informado qual base vai ficar na esquerda e na direita
# da "enderecos_sp" informo somente as colunas que me interessam (rua, cep...)
# o "how = left" fas um join entre as colunas informadas de cada coluna
dados_geo.head(3)
```

```python
dados_geo = dados_geo.drop_duplicates(subset=dados_vendas.columns).query('cep > 0')
# manter uma única linha para cada dado da tabela "dados_vendas"
# "drop_duplicates" remove as duplicadas relacionadas aos "dados_vendas"
dados_geo.head(3)
```
