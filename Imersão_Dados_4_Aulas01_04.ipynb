```python
dados_vendas['apenas_rua'] = dados_vendas['Rua'].str.extract(r'(^[\w ]+)') # esse regex descarta o ", num" de cada valor da coluna "Rua"
dados_vendas['apenas_rua'] = dados_vendas['apenas_rua'].str.lower().str.strip() # deixa tudo em min√∫sculo e tira espa√ßos extras
#isso deixa os endere√ßos com o mesmo padr√£o de formata√ß√£o da base "enderecos_sp"
dados_vendas.head(3)
```

```python
dados_geo = dados_geo.drop_duplicates(subset=dados_vendas.columns).query('cep > 0')
# manter uma √∫nica linha para cada dado da tabela "dados_vendas"
# "drop_duplicates" remove as duplicadas relacionadas aos "dados_vendas"
dados_geo.head(3)
```
